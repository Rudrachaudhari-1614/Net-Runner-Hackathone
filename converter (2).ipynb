```python
import os
import fitz  # PyMuPDF
import docx
from PIL import Image
import pytesseract
from bs4 import BeautifulSoup
import tifffile

# Function to convert PDF to text
def pdf_to_text(pdf_file):
    text = ""
    with fitz.open(pdf_file) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to convert DOCX to text
def docx_to_text(docx_file):
    text = ""
    doc = docx.Document(docx_file)
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

# Function to convert image (PNG, JPEG, TIFF) to text
def image_to_text(image_file):
    text = ""
    image = Image.open(image_file)
    text = pytesseract.image_to_string(image)
    return text

# Function to convert HTML to text
def html_to_text(html_file):
    with open(html_file, "r", encoding="utf-8") as file:
        soup = BeautifulSoup(file, "html.parser")
        text = soup.get_text()
    return text

# Function to convert TXT to text
def txt_to_text(txt_file):
    with open(txt_file, "r") as file:
        text = file.read()
    return text

# Function to determine file type and convert to text
def convert_to_text(file_path):
    file_extension = file_path.split(".")[-1].lower()
    if file_extension == "pdf":
        return pdf_to_text(file_path)
    elif file_extension == "docx":
        return docx_to_text(file_path)
    elif file_extension in ["jpg", "jpeg", "png", "tiff"]:
        return image_to_text(file_path)
    elif file_extension == "html":
        return html_to_text(file_path)
    elif file_extension == "txt":
        return txt_to_text(file_path)
    else:
        return "Unsupported file format."

# Take file path as input from the user
file_path = input("Enter the file path: ")

# Check if the file exists
if not os.path.exists(file_path):
    print("File not found.")
else:
    # Convert the file to text
    text = convert_to_text(file_path)
    print(text)
```
